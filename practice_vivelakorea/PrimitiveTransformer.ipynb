```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from pymks import (
    plot_microstructures,
    PrimitiveTransformer,
)
```

```python
im = Image.open('AlScEr_bin.png')
plt.figure()
plt.imshow(im, cmap='gray')
plt.title('sample')
plt.show()
```

```python
imarray = np.expand_dims(np.array(im), axis=0)
```

```python
print(imarray.shape)
print(imarray)
print(np.sum(imarray[0]) == 1034*1386)
```

```python
imarray = np.where(imarray == 255, 1, 0)
data = PrimitiveTransformer().transform(imarray)
print(data.shape)
plt.figure()
plt.imshow(data[0,:,:,0])
plt.show()

print(np.asarray(data[0,:,:,1]))
print(sum(np.asarray(data[0,:,:,1])))
```
