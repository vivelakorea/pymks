```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from pymks import (
    plot_microstructures,
    PrimitiveTransformer,
)
```

```python
im = Image.open('sample.tiff')
plt.figure()
plt.imshow(im, cmap='gray')
plt.title('sample')
plt.show()
```

```python
imarray = np.expand_dims(np.array(im), axis=0)
```

```python
print(imarray.shape)
mean = np.mean(imarray[0])
std = np.std(imarray[0])
print(mean, std)
```

```python
imarray[0] = np.array(imarray[0] - mean)
print(imarray[0])
imarray[0] = np.where(imarray[0] > 0, 1, 0)
print(imarray[0])
print(imarray.sum())
print(imarray[0].shape[0]*imarray[0].shape[1])
```

```python
data = PrimitiveTransformer().transform(imarray)
print(data.shape)
plt.figure()
plt.imshow(data[0,:,:,0])
plt.show()

print(np.asarray(data[0,:,:,1]))
```
